##### Description of cosine law

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
angle_range=np.linspace(-np.pi/2, np.pi/2, 1000)

def cosine_law(B=0):
    x = abs(np.cos(angle_range)*(1 + B*np.cos(angle_range)**2))*np.sin(angle_range)
    y = abs(np.cos(angle_range)*(1 + B*np.cos(angle_range)**2))*np.cos(angle_range)
    plt.plot(x, y)
```

```python
plt.figure(figsize=(6, 6))
plt.xlim(-0.75, 0.75)
plt.ylim(0, 1.5)
cosine_law(-0.6)
cosine_law(0)
cosine_law(0.3)
```
